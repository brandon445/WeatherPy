```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from api_key import api_key
import getpass

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
# Save config information
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"
```

# Perform API Calls

```python
# cityName = []
# cloudiness = []
# country = []
# date = []
# humidity = []
# lat = []
# long = []
# maxTemp = []
# windSpeed = []

# Get weather data
# weather_response = requests.get(query_url)
# weather_json = weather_response.json()

recordCounter = 0
setCounter = 1
citiesList = []

print("Beginning Data Retrieval")
print("-----------------------------")

for city in cities:
    query_url = f"{url}appid={api_key}&units={units}&q="
    response = requests.get(query_url + city).json()
    recordCounter = recordCounter + 1
    
    try:
        print(f"Processing Record " + str(recordCounter) + " of Set " + str(setCounter) + " | " + city)
        city = response["name"]
        cloudiness = response["clouds"]["all"]
        country = response["sys"]["country"]
        date = response["dt"]
        humidity = response["main"]["humidity"]
        lat = response["coord"]["lat"]
        long = response["coord"]["lon"]
        maxTemp = response["main"]["temp_max"]
        windSpeed = response["wind"]["speed"]
        
        citiesList.append({"City": city,
                           "Cloudiness": cloudiness,
                           "Country": country,
                           "Date": date,
                           "Humidity": humidity,
                           "Lat": lat,
                           "Long": long,
                           "Max Temp": maxTemp,
                           "Wind Speed": windSpeed
         })
        
        
        
    except:
        print("City not found.   Skipping...")
        
    #This is supposed to reset the Record count back to 0 once it hits 50 but it resets to 1
    if recordCounter == 50:
        recordCounter = 0
        setCounter = setCounter + 1

print("-----------------------------")
print("Data Retrieval Complete")
print("-----------------------------")
```

```python
citiesData = pd.DataFrame(citiesList)
citiesData.head()
```

```python
latitude = citiesData["Lat"]
maxTemp = citiesData["Max Temp"]
```
